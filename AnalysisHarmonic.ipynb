```python
def plot_gh_and_pos(exp):
    results = []
    fairness_ech_comm = []
    for elem in exp['experiments']:
        res = [
            elem['fair_set_size'],
            elem['group_harmonic'],
            elem['PoF'],
            elem['execution_time'],
            elem['fair_group_centrality_time']
        ]

        fec = [x[1][0] for x in elem['fair_harmonic_centrality'].items()

        ]
    
    
    
        results.append(res)
        fairness_ech_comm.append(fec)
    
    
    fig, axs = plt.subplots(2,figsize=(12, 8))
    n = exp['parameters']['n']
    p = exp['parameters']['p']
    fig.suptitle('Fair Group Harmonic Centrality n:'+str(n)+'   p:'+str(p))
    x = df['set_size'].values
    y1 = df['group_harmonic'].values
    y2  = df['PoF'].values
    axs[0].plot(x, y1,color = 'blue')
    axs[0].set_ylabel("Group Harmonic Centrality")
    axs[0].legend(['Group Harmonic Centrality'])
    axs[0].tick_params(
                        axis='x',          # changes apply to the x-axis
                        which='both',      # both major and minor ticks are affected
                        bottom=False,      # ticks along the bottom edge are off
                        top=False,         # ticks along the top edge are off
                        labelbottom=False) # labels along the bottom edge are off
    axs[1].plot(x, y2,color = 'coral')
    axs[1].set_ylabel("Price of Fairness")
    axs[1].set_xlabel("Fair Set Size")
    axs[1].legend(['PoF'])
    axs[1].set_xticks(x)
    plt.subplots_adjust(left=0.125,
                        bottom=0.1, 
                        right=0.9, 
                        top=0.9, 
                        wspace=1, 
                        hspace=0.0)
```
